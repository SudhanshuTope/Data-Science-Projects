```python
from tkinter import *
import tkinter as tk
from tkinter import ttk
import mysql.connector
import tkinter.messagebox
```

```python
class Database:
    def conn(self):
        print("Database: connection method called")
        con= mysql.connector.connect(host="localhost", user="root", password="#", database="inventory")
        cur = con.cursor()
        query= "CREATE TABLE IF NOT EXISTS Product (product_code integer PRIMARY KEY, product_name TEXT, product_qty TEXT,mfg_company TEXT, Inventory_Type TEXT )"
        cur.execute(query)
        con.commit()
        con.close()
        print("Connection Established")  
        
    def insert(self,prodno_var,prodname_var,prodqty_var,mfgcomp_var,invtype_var):
        print("Database: insert method called")
        con = mysql.connector.connect(host="localhost", user="root", password="#", database="inventory")
        cur = con.cursor()
        query = "INSERT INTO Product Values(%s,%s,%s,%s,%s)"
        cur.execute(query,(prodno_var,prodname_var,prodqty_var,mfgcomp_var,invtype_var))
        con.commit()
        con.close()
        print("Database: finished")

    def show(self):
        print("Database: show method called")
        con = mysql.connector.connect(host="localhost", user="root", password="#", database="inventory")
        cur = con.cursor()
        query= "SELECT * FROM Product"
        cur.execute(query)
        rows= cur.fetchall()
        con.close()
        print("Database finished\n")
        return rows
    
    def delete(self,prodno_var):
        print("Database: delete method called",prodno_var)
        con = mysql.connector.connect(host="localhost", user="root", password="#", database="inventory")
        cur = con.cursor()
        cur.execute("DELETE FROM Product where product_code=%s",(prodno_var,))
        con.commit()
        con.close()
        print(prodno_var,"Datbase finished\n")
        
    def search(self,prodno_var="",prodname_var="",prodqty_var="",mfgcomp_var="",invtype_var=""):
        print("Database: search method called")
        con = mysql.connector.connect(host="localhost", user="root", password="#", database="inventory")
        cur = con.cursor()
        cur.execute("SELECT * FROM Product where product_code=%s or product_name=%s or product_qty=%s or mfg_company=%s or Inventory_Type=%s",(prodno_var,prodname_var,prodqty_var,mfgcomp_var,invtype_var))
        row= cur.fetchall()
        con.close()
        print(prodno_var,"Database finished\n")
        return row
    
    def update(self,prodno_var="",prodname_var="",prodqty_var="",mfgcomp_var="",invtype_var=""):
        print("Database: search method called")
        con = mysql.connector.connect(host="localhost", user="root", password="#", database="inventory")
        cur = con.cursor()
        cur.execute("UPDATE Product set product_code=%s or product_name=%s or product_qty=%s or mfg_company=%s or Inventory_Type=%s where product_code=%s",(prodno_var,prodname_var,prodqty_var,mfgcomp_var,invtype_var))
        con.commit()
        con.close()
        print(prodno_var,"databse finished\n")
        
```

```python
class Product:
    def __init__(self,root):
        p=Database()
        p.conn()
        
        self.root=root
        self.root.title("Warehouse Inventory Management System")
        self.root.geometry("1350x750+0+0")
        
# Command Functions
        def clear():
            print("product: clear methoid called")
            self.productno_box.delete(0, END)
            self.productname_box.delete(0, END)
            self.productqty_box.delete(0, END)
            self.mfgcomp_box.delete(0, END)
            self.invtype_box.delete(0, END)
            print("product: clear finished")

        def insert():
            print("product: insert method starts")
            if (len(prodno_var.get())!=0):
                p.insert(prodno_var.get(),prodname_var.get(),prodqty_var.get(),mfgcomp_var.get(),invtype_var.get())
                self.Table.delete(0,END)
                self.Table.insert(END,prodno_var.get(),prodname_var.get(),prodqty_var.get(),mfgcomp_var.get(),invtype_var.get())
                showilist()
            else:
                tkinter.messagebox.askyesno("WAREHOUSE INVENTORY STORAGE MANAGEMENT SYSTEM","Enter Product Code")
            print("Product: insert method finished")

        def showilist():
            print("product: show method ")
            self.Table.delete(0,END)
            for row in p.show():
                self.Table.insert(END, row, str(""))
            print("product: show prduct finished\n")

        def prodrec(event):
            print("product: prodrec called")
            global pd
            searchpd = self.Table.curselection()[0]
            pd= self.Table.get(searchpd)
            self.productno_box.delete(0,END)
            self.productno_box.insert(END, pd[0])

            self.productname_box.delete(0, END)
            self.productname_box.insert(END, pd[1])

            self.productqty_box.delete(0, END)
            self.productqty_box.insert(END, pd[2])

            self.mfgcomp_box.delete(0, END)
            self.mfgcomp_box.insert(END, pd[3])

            self.invtype_box.delete(0, END)
            self.invtype_box.insert(END, pd[4])

            print("product: prodrec finished")

        def delete():
            print("product: delete starts")
            if (len(prodno_var.get())!=0):
                p.delete(pd[0])
                clear()
                showilist()
            print("prodct: delete finished")

        def search():
            print("product: search method")
            self.Table.delete(0,END)
            for row in p.search(prodno_var.get(),prodname_var.get(),prodqty_var.get(),mfgcomp_var.get(),invtype_var.get()):
                self.Table.insert(END, row, str(""))
            print("product: search finished")

        def update():
            print("product: update called")
            if (len(prodno_var.get()!=0)):
                print("pd[0]",pd[p])
                p.delete(pd[0])
            if(len(prodno_var.get()!=0)):
                p.insert(prodno_var.get(),prodname_var.get(),prodqty_var.get(),mfgcomp_var.get(),invtype_var.get())
                self.Table.delete(0,END)
            self.Table.insert(END,(prodno_var.get(),prodname_var.get(),prodqty_var.get(),mfgcomp_var.get(),invtype_var.get()))
            print("product: update finished")
#Frames        
        Frame_Title=tk.Label(self.root,text="Warehouse Management System",bd=10,relief=GROOVE,font=("times new roman",40,"bold"),bg="yellow",fg="red")
        Frame_Title.pack(side=TOP,fill=X)
        
        Frame1=Frame(self.root,bd=4,relief=RIDGE,bg="cyan")
        Frame1.place(x=20,y=100,width=450,height=560)
        
        Frame2=Frame(self.root,bd=4,relief=RIDGE,bg="cyan")
        Frame2.place(x=500,y=100,width=1000,height=560)
        
        #listing_frame= LabelFrame(self.root, bd=4, padx=50, pady=10, bg="cyan", relief=RIDGE, font=("arial", 13, "bold"), text="PRODUCT ITEM INFORMATION:")
        #listing_frame.place(x=500,y=100,width=1000,height=560)
        
        Frame_button=Frame(Frame1,bd=4,relief=RIDGE,bg="cyan")
        Frame_button.place(x=10,y=400,width=420,height=130)
        
        listing_frame = Frame(Frame2,bd=4, relief=RIDGE,bg='cyan')
        listing_frame.place(x=10,y=70,width=950,height=470)

        
#Labels
        title1=tk.Label(Frame1,text="Inventory Widget",bg="cyan",fg="black",font=("times new roman",20,"bold"))
        title1.grid(row=0,columnspan=2,pady=20)
        
        self.product_no=tk.Label(Frame1,text="Product Code",bg="cyan",fg="black",font=("times new roman",15,"bold"))
        self.product_no.grid(row=1,column=0,pady=10,padx=20,sticky='w')
        prodno_var=tk.StringVar()
 
        self.product_name=tk.Label(Frame1,text="Product Name",bg="cyan",fg="black",font=("times new roman",15,"bold"))
        self.product_name.grid(row=2,column=0,pady=10,padx=20,sticky='w')
        prodname_var=tk.StringVar()
        
        self.product_quantity=tk.Label(Frame1,text="Product Quantity",bg="cyan",fg="black",font=("times new roman",15,"bold"))
        self.product_quantity.grid(row=3,column=0,pady=10,padx=20,sticky='w')
        prodqty_var=tk.StringVar()
        
        self.Mfg_company=tk.Label(Frame1,text="Mfg.Company",bg="cyan",fg="black",font=("times new roman",15,"bold"))
        self.Mfg_company.grid(row=4,column=0,pady=10,padx=20,sticky='w')
        mfgcomp_var=tk.StringVar()
        
        self.inv_type=tk.Label(Frame1,text="Inventory Type",bg="cyan",fg="black",font=("times new roman",15,"bold"))
        self.inv_type.grid(row=5,column=0,pady=10,padx=20,sticky='w')
        invtype_var=tk.StringVar()
        
        self.title2=tk.Label(Frame2,text="Inventory Details",bg="cyan",fg="black",font=("times new roman",20,"bold"))
        self.title2.grid(row=0,columnspan=2,pady=20)
    
    
        
#Entry Box 
        self.productno_box=tk.Entry(Frame1,font=("times new roman",15,"bold"),bd=5,relief=GROOVE,textvariable=prodno_var)
        self.productno_box.grid(row=1,column=1,pady=10,padx=10,sticky='w')
        
        self.productname_box=tk.Entry(Frame1,font=("times new roman",15,"bold"),bd=5,relief=GROOVE,textvariable=prodname_var)
        self.productname_box.grid(row=2,column=1,pady=10,padx=10,sticky='w')
        
        self.productqty_box=tk.Entry(Frame1,font=("times new roman",15,"bold"),bd=5,relief=GROOVE,textvariable=prodqty_var)
        self.productqty_box.grid(row=3,column=1,pady=10,padx=10,sticky='w')
        
        self.mfgcomp_box=tk.Entry(Frame1,font=("times new roman",15,"bold"),bd=5,relief=GROOVE,textvariable=mfgcomp_var)
        self.mfgcomp_box.grid(row=4,column=1,pady=10,padx=10,sticky='w')
        
        self.invtype_box=ttk.Combobox(Frame1,font=("times new roman",15,"bold"),state='readonly',textvariable=invtype_var)
        self.invtype_box['values']=('Select','Raw Material','Finished Goods','MRO Inventory','Packing Material','Merchandise Inventory','Manufacturing Inventory')
        self.invtype_box.grid(row=5,column=1,pady=10,padx=10,sticky='w')
        
        
        

#Buttons
        self.Add_button1=tk.Button(Frame_button,text='Save',width=25,height=2,command=insert)
        self.Add_button1.grid(row=0,column=0,padx=10)
        
        self.Add_button2=tk.Button(Frame_button,text='Show Data',width=25,height=2,command=showilist)
        self.Add_button2.grid(row=0,column=1,padx=10)
        
        self.Add_button3=tk.Button(Frame_button,text='Clear',width=25,height=2,command=clear)
        self.Add_button3.grid(row=1,column=0,padx=10)
        
        self.Add_button4=tk.Button(Frame_button,text='Update',width=25,height=2,command=update)
        self.Add_button4.grid(row=1,column=1,padx=10)
        
        self.Add_button5=tk.Button(Frame_button,text='Delete',width=25,height=2,command=delete)
        self.Add_button5.grid(row=2,column=0,padx=10)
        
        self.Add_button6=tk.Button(Frame_button,text='Search',width=25,height=2,command=search)
        self.Add_button6.grid(row=2,column=1,padx=10)
        
        scroll_x = Scrollbar(listing_frame,orient=HORIZONTAL)
        scroll_y = Scrollbar(listing_frame,orient=VERTICAL)
        #self.Table=ttk.Treeview(listing_frame,columns=("Product Code","Product Name","Product Qty","Mfg. Company","Inventory Type"),xscrollcommand=scroll_x.set,yscrollcommand=scroll_y.set)
        self.Table= Listbox(listing_frame,width=40, height= 16, font=("arial",15,"bold"), xscrollcommand=scroll_x.set,yscrollcommand=scroll_y.set)
        self.Table.bind('<<ListboxSelect>>',prodrec)
        #self.Table.grid(row=0,column=1,padx=8)
        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=RIGHT,fill=Y)
        scroll_x.config(command=self.Table.xview)
        scroll_y.config(command=self.Table.yview)
        

 
        self.Table.pack(fill=BOTH,expand=1)  
        
if __name__ == '__main__':
    root= Tk()
    ob = Product(root)
    root.mainloop()        

        
```
